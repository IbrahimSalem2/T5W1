## 1. Exploratory Data Analysis (EDA)

### Perform summary statistics on the dataset.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Traffic_2.csv')
```

```python


df['Date'] = pd.to_datetime(df['Datetime']).dt.date
df['Time'] = pd.to_datetime(df['Datetime']).dt.time
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
# prompt: drop rows from a colomn contains nulls
T1=df.copy()
T1 = df.dropna(subset=['TravelTimeHistoric'])
T1 = df.dropna(subset=['TravelTimeLive'])
df=T1.copy()
```

```python
T1.isnull().sum()
```

### Identify and analyze patterns in the data.

```python
df_corr = df[['TrafficIndexLive', 'JamsCount', 'JamsDelay', 'JamsLength', 'TrafficIndexWeekAgo', 'TravelTimeHistoric', 'TravelTimeLive']]
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(df_corr.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

### Visualize the distribution of key variables (e.g., Traffic_Index, Date).

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['TrafficIndexLive'], bins=25, kde=True, color='red')
plt.title('Distribution of Traffic Index')
plt.xlabel('Traffic Index')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['TrafficIndexWeekAgo'], bins=25, kde=True, color='blue')
plt.title('Distribution TrafficIndexWeekAgo   ')
plt.xlabel('TrafficIndexWeekAgo    Index')
plt.ylabel('Frequency')
plt.show()
```

### Explore relationships between variables (e.g., Traffic_Index vs. Weather_Condition).


```python

org=df.copy()

df['Month'] = pd.to_datetime(df['Date']).dt.month


monthlyAvgtTraffic= df.groupby('Month')['TrafficIndexLive'].mean()


plt.figure(figsize=(10, 6))
plt.bar(monthlyAvgtTraffic.index, monthlyAvgtTraffic.values)
plt.title('Average Traffic Index by Month')
plt.xlabel('Month')
plt.ylabel('Average Traffic Index')
plt.show()
```

```python

df['Month'] = pd.to_datetime(df['Date']).dt.month


monthlyAvgtTraffic= df.groupby('City')['TravelTimeLive'].mean()


plt.figure(figsize=(10, 6))
plt.bar(monthlyAvgtTraffic.index, monthlyAvgtTraffic.values)
plt.title('Average TravelTimeLive by City')
plt.xlabel('City')
plt.ylabel('Average Traffic Index')
plt.show()
```
